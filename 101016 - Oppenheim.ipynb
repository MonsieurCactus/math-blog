```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 5,5
```

```python
a,b,c = 1, 1, -1*np.sqrt(2)
```

```python
N = 15
n = np.arange(0, N)
```

```python
Q = a*n[None,None,...]**2 + b*n[None,...,None]**2 + c*n[...,None,None]**2

Q = np.sort(Q.flatten())
```

```python
np.max(Q), np.min(Q)
```

```python
t = np.arange( 1.05*np.min(Q)//1 , 1.05*np.max(Q)//1 , 0.1)
x = 0*t
```

```python
y = 0*t + 200

for q in Q:
    y = np.min( np.vstack((np.abs(t - q), y)), axis=0)
    
```

```python
plt.plot(t, y, '.', markersize=2)
```

```python
plt.plot(t[ (np.abs(t) < 10)], y[ (np.abs(t) < 10)], '-', markersize=2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 7,7
```

```python
import time
import numpy as np

t = time.time()

N = 1000000
x = np.zeros(N).astype(int)
x[0] = 1
x[1] = 1

p = 0
while( p < np.sqrt(N)):
    p = p + 1
    while(x[p] == 1):
        p += 1
    x[2*p::p] = 1
    
P = np.where(x == 0)[0]

for k in 1+np.arange(100-1):
    print k,
    x = np.zeros(N).astype(int)
    for p in P[P < np.sqrt((k+1)*N)]:
        if( (k*N)% p == 0):
            x[0] = 1
        x[ p - (k*N % p)::p] = 1

    P = np.hstack((P, np.where(x == 0)[0] + k*N))

print time.time() - t
P.shape
```

```python
p = P[50]

z = []

sq = {t**2 % p :t for t in range( (p-1)/2)}

#print p
#print sq


for A in range(p):
    print A,
    
    z += [[]]
    for B in range(p):
        z[A] += [[(x,  1*sq[(x**3 + A*x + B)%p]) for x in range(p)  if (x**3 + A*x + B)%p in sq]]
        #z[A] += [[(x, -1*sq[(x**3 + A*x + B)%p]) for x in range(p)  if (x**3 + A*x + B)%p in sq]]
```

```python
plt.hist([2*len(z[A][B])-p-1  for A in range(p) for B in range(p)], bins = -50 + np.arange(200) )
plt.show()
```

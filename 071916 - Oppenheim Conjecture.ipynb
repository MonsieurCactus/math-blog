```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,6
```

```python


for t in np.arange(10):
    O = []
    for N in np.arange(200)+2:

        x = np.arange(N)

        a = 1
        b = t*0.1+np.sqrt(2)

        z = np.sqrt((x[...,None]**2 + a*x[None,...]**2 )/b)

        X = x[...,None]**2 + a*x[None,...]**2 - b*np.floor(z).astype(int)**2
        Y = x[...,None]**2 + a*x[None,...]**2 - b*np.ceil(z).astype(int)**2

        O += [np.sort(np.minimum(X,-Y).reshape(N**2))[1]]
    
    plt.plot(O, color=str(t*0.1), alpha=1)
    
N = np.arange(200)+2
plt.plot(N, N**-0.4, linewidth=5)

plt.xlim([0,200])
plt.title("minimum of $|x^2 + y^2 - bz^2|$ for $b \in [\sqrt{2}, \sqrt{2}+1]$", fontsize=20)
plt.xlabel("<--- b --->", fontsize=20)
plt.ylabel("$\min |Q(x)|$ for $x \in \mathbb{Z}^3$", fontsize=20)
```

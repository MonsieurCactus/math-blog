# An Approach to the Oppenheim Conjecture

* http://arxiv.org/abs/1604.02413 **Small gaps in the spectrum of the rectangular billiard**
* http://arxiv.org/abs/1604.02087 **A quantitative Oppenheim Theorem for generic diagonal quadratic forms**
    
What is the smallest value you can obtain for $x^2 + y^2 - \sqrt{2}z^2$ as $(x,y,z)$ varies over the integers $\mathbb{Z}$?

If you replace $\sqrt{2}$ with any irrational number, this value tends to $0$.  New papers address special cases of Theorem

```python


for t in np.arange(0, 100):
    print t,
    O = []
    for N in np.arange(200)+2:

        x = np.arange(N)

        a = 1
        b = t*0.01+np.sqrt(2)

        z = np.sqrt((x[...,None]**2 + a*x[None,...]**2 )/b)

        X = x[...,None]**2 + a*x[None,...]**2 - b*np.floor(z).astype(int)**2
        Y = x[...,None]**2 + a*x[None,...]**2 - b*np.ceil(z).astype(int)**2

        O += [np.sort(np.minimum(X,-Y).reshape(N**2))[1]]
    
    plt.plot(O, color=str(t*0.01), alpha=1)
    
N = np.arange(200)+2
plt.plot(N, N**-0.4, linewidth=5)
plt.plot(N, N**-1, linewidth=5, color='r')

plt.xlim([0,200])
plt.title("minimum of $|x^2 + y^2 - bz^2|$ for $b \in [\sqrt{2}, \sqrt{2}+1]$", fontsize=20)
plt.xlabel("<--- N --->", fontsize=20)
plt.ylabel("$\min |Q(x)|$ for $x \in \mathbb{Z}^3$", fontsize=20)
```

```python
for t in np.arange(0, 100):
    print t,
    O = [1]

    for N in np.arange(200)+2:

        a = 1
        b = t*0.01+np.sqrt(2)

        x = np.arange(N+1)

        z = np.sqrt( (N**2 + a*x**2 )/b )

        X1 = N**2 + a*x**2 - b*np.floor(z)**2
        X2 = N**2 + a*x**2 - b*np.ceil(z)**2

        z = np.sqrt( (x**2 + a*N**2 )/b )

        Y1 = x**2 + a*N**2 - b*np.floor(z)**2
        Y2 = x**2 + a*N**2 - b*np.ceil(z)**2

        w = np.min( [np.min(X1), np.min(-X2), np.min(Y1), np.min(-Y2)] )

        O += [np.minimum(O[-1], w)]

        plt.plot(O, color=str(t*0.01), alpha=1, linewidth=0.5)
        
N = np.arange(200)+2
plt.plot(N, N**-0.4, linewidth=5)
plt.plot(N, N**-1, linewidth=5, color='r')
plt.xlim([0,200])
```

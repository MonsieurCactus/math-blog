# Hardy's Curious Result


We want to know about what happens to the fraction part as the wind around the clock and stop at regular intervals.  Not 1/4 or 1/8 but something irregular like $1/\sqrt{3}$.

$$ \Large \sum_{\nu \leq n } \{ \nu \theta\}^2 = o(n) \hspace{0.25in}\text{   possibly  }\hspace{0.25in} O(1)$$

Hardy's big-O notation is a nuisiance.  Here $\{ x\} = [x] - (x) $ e.g. $\{ 2.25\} = 0.25, \{-1.3\} = -0.3$

```python
n = np.arange(100000)
theta = np.sqrt(5)

x = (n*theta) % 1 - 0.5

plt.plot(np.cumsum(x), '.', markersize=1)
```

$$ \Large \sum_{\nu \leq n } \{ \nu \theta\}^2 = \frac{1}{12}n + O(1) $$

```python
n = np.arange(50000)
theta = np.sqrt(7)

x = (n*theta) % 1 - 0.5

plt.plot(n, np.cumsum(x**2), '.', markersize=1)
```

$$ \Large \sum_{\nu \leq n } \{ \nu \theta\}^2 - \frac{1}{12}n = O(1) $$

```python
n = np.arange(50000)
theta = np.sqrt(19)

x = (n*theta) % 1 - 0.5

plt.plot(n, np.cumsum(x**2) - n*1.0/12, '.', markersize=1)
```

# Unenlightening Histogram of Error Terms

```python
n = np.arange(50000)
theta = np.sqrt(7)

x = (n*theta) % 1 - 0.5

plt.hist( np.cumsum(x**2) - n*1.0/12, bins=100)
```

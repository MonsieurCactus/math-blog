```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 15,15
import time
```

```python
def SOS(M):

    N = int(M**0.5)

    z = [ (a,b,c) for a in range(N) for b in range(N) for c in range(N) if a**2 + b**2 + c**2 == M]
    z = np.array(z)
    
    return z
```

```python
def g(n):
    if n == 0:
        return False
    elif n%4 == 0:
        return g(n/4)
    elif n%4 == 2:
        return True
    elif n%8 == 7:
        return False
    else:
        return True
    
print [n for n in range(100) if not g(n)]
```

```python
# is there a tool for turning Python into flow charts?

A,B = 12000, A+100

start_time = time.time()

for M in range(A,B):
    
    #print M,
    
    if g(M):
        z = SOS(M)
    else:
        z = np.array([])
    #print z.shape
    
    #print time.time() - start_time
    start_time = time.time()

    if z.shape[0] > 0:
        for a in [-1,1]:
            for b in [-1,1]:
                plt.plot(  a*z[:,0]/M**0.5/2.5 + 1*(M%10) + 0.5 
                         , b*z[:,1]/M**0.5/2.5 + 1*(M//10 - A/10)+ 0.5
                         , 'k.'
                         , markersize= 2)
                
plt.xlim([0, 10])
plt.ylim([0, 10])

plt.grid(True)

plt.xticks( [2*x for x in range(5)], ["" for a in range(10)])
plt.yticks( [2*x for x in range(5)], ["" for a in range(10)])

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,8
```

```python
N = 10**3
n = np.arange(N) + 1

# name a dense set of numbers
a = 1.0/n**2
X = (n*3**0.5)%1

t = np.arange(0,1,10**-5)
z = 0*t

for n in np.arange(N):
    z[(t > X[n])] += (1.0/(n+1)**2)
    
plt.plot(t,z, 'b.', markersize=0.01)
plt.grid(True)
```

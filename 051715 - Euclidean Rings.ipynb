```python
L = 5
d = 1

a = 10+5j
b = 1+4j

z = np.array([0,1,1+1j, 1j])
z = b*z
plt.fill( z.real, z.imag, '#F0F0F0')

x = np.arange(-L, L+2)
z = x[...,None]+1j*x[None,...]
z = a - b*z
plt.plot(z.real, z.imag, '-', c='#A0A0A0')

x = np.arange(-L, L+2)
z = x[None,...]+1j*x[...,None]
z = a - b*z
plt.plot(z.real, z.imag, '-', c='#A0A0A0')

x = np.arange(-L, L+1)
z = x[...,None]+1j*x[None,...]
z = a - b*z

plt.plot(z.real, z.imag, 'k.')


plt.xlim([-15,15])
plt.ylim([-15,15])
#plt.grid(True)
plt.show()
```

Euclidean algorithm says $a = bq + r$ with $|r| < b/2$.  Then $\mathbb{Q}(\sqrt{-d}) = \mathbb{Z}[\sqrt{-d}] + O$  

```python
d = 1

D = np.sqrt(d)*(d % 4 == 1) + 0.5*(1 + np.sqrt(d))*(d%4 == 3)

dt = 0.01
t  = 2*np.pi*np.arange(0,1+dt, dt)
z = 1.0/np.sqrt(2)*(np.cos(t) + 1j*np.sin(t)/np.sqrt(d))

L = 5
for m in range(-L,L):
    for n in range(-L,L):
        plt.fill(m + z.real, n + z.imag, alpha=0.2)

plt.axis("Equal")
```

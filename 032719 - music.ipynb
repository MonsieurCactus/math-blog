```python
# right channel
plt.plot(x[1][:,1])
```

```python
# length of audio (in seconds)
x[1].shape[0]/x[0]
```

```python
t  = 3.25
n  = int(48000*t)
dn = int(48000*(0.01))

plt.plot(x[1][n:n+dn,1])
```

```python
import numpy as np
```

```python
t  = 3.25
n  = int(48000*t)
dn = int(48000*(0.5))
```

```python
dn
```

```python
# https://docs.scipy.org/doc/numpy/reference/generated/numpy.fft.fft.html

sp = np.fft.fft(x[1][n:n+dn,1])
freq = np.fft.fftfreq(dn)

plt.plot(freq, sp.real, freq, sp.imag)
```

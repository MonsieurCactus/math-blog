```python
n = np.arange(200)+1
t = np.zeros(200)
t[::2] = 1
t[1::2] = -1

n0 = np.sqrt(n)*t

n1 = np.cumsum(n0)
#plt.plot(n1, '.', markersize=3)

n2 = np.cumsum(n1)
plt.plot(n2/n, '-', markersize=3)

n3 = np.cumsum(n2)
plt.plot(2*n3/n**2, '-', markersize=3)

n4 = np.cumsum(n3)
plt.plot(6*n4/n**3, '-', markersize=3)

plt.xlim([50,200])
plt.ylim([0.2,0.5])
```

```python
dx = 0.0025
x  = np.pi*np.arange(0-dx,1+dx,dx)

z = x[...,None] + x[None,...]

w = 0*z

A = np.random.normal(0,1,(2*L+2,2*L+2))

L = 50
for m in range(-L,L+1):
    for n in range(-L,L+1):
        #w += (2*np.random.random()-1)*np.sin(m*x[...,None])*np.sin(n*x[None,...])
        w += (A[m+L,n+L])*np.sin(m*x[...,None])*np.sin(n*x[None,...])

pos = (np.abs(w[1:,1:]>0) & np.abs(w[:-1,1:]>0)& np.abs(w[1:,:-1]>0)&np.abs(w[:-1,:-1]>0))
neg = (np.abs(w[1:,1:]<0) & np.abs(w[:-1,1:]<0)& np.abs(w[1:,:-1]<0)&np.abs(w[:-1,:-1]<0))

#sgn = 1*(pos|neg)
sgn =  (1*pos)
a,b = np.where(sgn == 0)

fig, ax = plt.subplots(dpi=72)

ax.plot(x[a], x[b], '.', markersize=2, color="#E8885A")
plt.xlim([-0.25,3.25])
plt.ylim([-0.25,3.25])
plt.axis("Off")
```

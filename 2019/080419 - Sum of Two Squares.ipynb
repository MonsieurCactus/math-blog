```python
a = 10**4
a = sympy.nextprime(a)

while a < 10**4 + 10**4:
    if (a %4 == 1):
        #print(a)
        z = [(x,y) for x in range(int(a**0.5)+1) for y in range(int(a**0.5)+1) if x**2 + y**2 - a == 0 and x < y ]
        #print(z)
        #if len(z) > 0:
        #    print("%s^2+%s%2"%(z[0][0], z[0][1]))
        #    print("%s×%s + %s×%s = %s"%(z[0][0], z[0][0], z[0][1],z[0][1], z[0][0]**2 + z[0][1]**2))
        z = [(x, int((a-x**2)**0.5)) for x in range(int(a**0.5)+1) if (a - x**2)**0.5 % 1 < 0.01 and (a - x**2)**0.5 % 1 > 0.0001 ]
        #print(z)
        if len(z) > 0:
        #    print("%s^2+%s%2"%(z[0][0], z[0][1]))
            #print(a)
            print("\\item $%s×%s + %s×%s = %s$"%(z[0][0], z[0][0], z[0][1],z[0][1], z[0][0]**2 + z[0][1]**2))
            print("\\item $\\sqrt{ {\color{green!50!black}{%s}} - %s\\times %s } \\approx [...] + %s$"%(a, z[0][0],z[0][0],np.round((a - z[0][0]**2)**0.5%1, decimals=5)))
        #print()
    a = sympy.nextprime(a)
```

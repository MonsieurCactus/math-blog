```python
vals = im.mean(axis=2).flatten()
```

```python
b, bins, patches = plt.hist(vals, range(256))
plt.xlim([0,255])
plt.show()
```

```python
color = ['b', 'g', 'r']

for i, col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])

plt.grid(True)
plt.show()
```

# Error Estimate

$$ \Big| \sum_{0 \leq m,n \leq N} \{ \sqrt{2} m + \sqrt{3} n \}^n - \int_0^1 x^n \, dx \Big| \stackrel{?}{<} \frac{1}{N^2} $$

```python
X = []

for N in 2+ np.arange(500):

    n = np.arange(N)

    x = (np.sqrt(2)*n[None,...] + np.sqrt(3)*n[...,None])%1

    t = x.reshape((N**2))

    X += [ np.sum(t**3) / N**2 - 0.25 ]
    
X = np.array(X)

plt.plot( 1+ np.arange(500) , np.log(np.abs(X))/np.log(2+ np.arange(500)) , '.' , markersize = 3 )
plt.grid(True)
```

# Error Estimate ($n = 5$ example)

$$ \Big| \sum_{0 \leq m,n \leq N} \{ \sqrt{2} m + \sqrt{3} n \}^5 - \int_0^1 f(x) \, dx \Big| \stackrel{?}{<} \frac{1}{N^2} $$

```python
X = []

for N in 2+ np.arange(1000):

    n = np.arange(N)

    x = (np.sqrt(2)*n[None,...] + np.sqrt(3)*n[...,None])%1

    t = x.reshape((N**2))

    X += [ np.sum(t**5) / N**2 - 1.0/6 ]
    
X = np.array(X)

plt.plot( 1+ np.arange(1000) , np.log(np.abs(X))/np.log(2+ np.arange(1000)) , '.' , markersize = 3 )
plt.grid(True)
```

# Simultaneous Quadratic Forms

two equations and two variables not enough to obtain a subset of $\mathbb{Z}^2$ with a density $>0$

two equations and **3**? variables?

```python
f = lambda a,b : a**2 + 2*b**2
g = lambda a,b : 2*a**2 + 2*b**2
```

```python
plt.rcParams['figure.figsize'] = 7,7
```

```python
x = [(f(a,b), g(a,b)) for a in range(600) for b in range(600)]
x = np.array(x)
plt.plot(x[:,0], x[:,1], '.', markersize=1)
plt.xlim([300000,400000])
plt.ylim([400000,500000])
plt.xticks([])
plt.yticks([])
```

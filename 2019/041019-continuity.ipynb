```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = 0.0001
t = np.arange(0,1,dt)
```

```python
f = lambda x : (x < 0.5)*0 + (x > 0.5)*1


z = 0*t

N = 100

for m in range(N):
    for n in range(N):
        z += f( (t - (m**2 + 2**0.5*n**2))%1 )
        
plt.plot(t,N**-2*z, '.', markersize=1)
plt.plot(t,0*t + 0.5)
```

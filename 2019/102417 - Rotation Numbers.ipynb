```python
c = np.arange(0,1,0.001)
    
x = np.arange(0,1,0.001)
y = np.arange(0,1,0.001)

def g(x):
    return x + 0.05*np.sin(8*np.pi*x)

# H(x+m) = H(x) + m

def H(x):
    #return x  - 1 * ( 0.25 - (x-0.5)**2 )
    return x*x*(1-x)*(1-x)

def h(x):
    #return H(x%1) + (x//1)
    return H(x%1)
    
for n in 1 + np.arange(10000):
    y = h(y + c) 
        
plt.plot(x, (y-x)/n , '.', color='k', alpha=0.5, markersize=1)

plt.xlim([-0.1, 1.1])
plt.ylim([-0.1, 1.1])

plt.grid(True)
```

```python
N = 10**5

x = (2*np.random.random((N,3)) - 1)

x = x/((np.sum(x**2, axis=1))**0.5)[...,None]

p = np.random.random(N)

f = x[:,0]**2 - 2*x[:,1]**2 + x[:,2]**2

f += 0.5*(x[:,1]**2 - 2*x[:,2]**2 + x[:,0]**2)

```

```python
# can we take the average of a function over a group orbit ?

A = np.random.random((3,3))

A[:,0] = A[:,0]/np.sum(A[:,0]**2)**0.5

A[:,1] = A[:,1] - np.dot(A[:,1], A[:,0])*A[:,0]

A[:,1] = A[:,1]/np.sum(A[:,1]**2)**0.5

A[:,2] = A[:,2] - np.dot(A[:,2], A[:,1])*A[:,1] - np.dot(A[:,2], A[:,0])*A[:,0]

A[:,2] = A[:,2]/np.sum(A[:,2]**2)**0.5

np.dot(A, A.T)
```

```python
z = np.dot(x,A)

mask = (f > 0)

plt.plot(z[mask][:,0], z[mask][:,1], 'k.', markersize=0.15)

plt.plot(z[:,0], z[:,1], 'b.', markersize=0.05)

plt.grid(True)
```

```python
# try to generate a uniform spanning tree of the rectangular grid 

V = []
E = []

# not visited vertices
L  = 25
V1 = [ a+1j*b for a in range(-L-1,L+1) for b in range(-L-1,L+1) if a**2 + b**2  < (L+0.5)**2] 

x = 0
V += [x]

print("abc")

for t in range(len(V1)-1):
    


    dx = [1, -1, 1j, -1j]

    #print(len([v for v in V1 if v not in V]), len(V1), len(V))
    
    x = [v for v in V1 if v not in V][int((len(V1) - len(V))*np.random.random())]

    while x not in V:

        z = x + dx[int(4*np.random.random())]

        if np.abs(z) < L + 0.5:

            if z in V:
                E += [(x,z)]
                
                V += [x]

            x = z
        
#E, V
```

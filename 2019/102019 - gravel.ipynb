```python
# superimpose two grids and only draw the lines that fit in a circle

# pick two diametrically opposite points


dt = 0.1
ds = 0.1*np.random.random()
T = np.arange(-1+ds,1,dt)

theta = np.random.random()
for t in T:
    z = t - 1j*(1-t**2)**0.5
    w = t + 1j*(1-t**2)**0.5
    
    z *= np.exp(2j*np.pi*theta)
    w *= np.exp(2j*np.pi*theta)
    
    plt.plot([z.real, w.real], [z.imag, w.imag], 'k-', linewidth=1)
    
    
for t in T:
    z = t - 1j*(1-t**2)**0.5
    w = t + 1j*(1-t**2)**0.5

    z *= np.exp(2j*np.pi*(theta+0.25))
    w *= np.exp(2j*np.pi*(theta+0.25))
    
    plt.plot([z.real, w.real], [z.imag, w.imag], 'k-', linewidth=1)
    
dt = 0.2
ds = 0.1*np.random.random()
T = np.arange(-1+ds,1,dt)

theta = np.random.random()
for t in T:
    z = t - 1j*(1-t**2)**0.5
    w = t + 1j*(1-t**2)**0.5

    z *= np.exp(2j*np.pi*theta)
    w *= np.exp(2j*np.pi*theta)
    
    plt.plot([z.real, w.real], [z.imag, w.imag], 'k-', linewidth=2)
    
for t in T:
    z = t - 1j*(1-t**2)**0.5
    w = t + 1j*(1-t**2)**0.5

    z *= np.exp(2j*np.pi*(theta+0.25))
    w *= np.exp(2j*np.pi*(theta+0.25))
    
    plt.plot([z.real, w.real], [z.imag, w.imag], 'k-', linewidth=2)
```

```python
# 100 random points in a square minimum distance D apart.

z = np.random.random() + 1j*np.random.random()
P = [z]
d = 0.05

while len(P) < 250:
    z = np.random.random() + 1j*np.random.random()
    if np.min([ np.abs(z - p) for p in P]) > d:
        P += [z]
    
P = np.array(P)


t = np.arange(0,1.01,0.01)
z = np.exp(2j*np.pi*t)

for p in P:
    r = 0.5*d*(np.random.random() + 1)
    Q = p + r*z
    plt.plot(Q.real, Q.imag, 'k-', linewidth=0.5)
    
plt.plot(P.real, P.imag, 'k.', markersize=1)
plt.grid(True)
```

```python
N = 100
m = np.arange(N)
Z = (m[None,...]**2 + 2**0.5*m[...,None]).reshape(N**2) 

t = np.arange(0,N**2,1)

x = np.zeros(N**2)

for z in Z:
    x[(t > z)] += 1

plt.plot(t,x, '.')

plt.plot(t, 0.5*(t-10**4)+10**4)
```

```python
X = [(m,n) for m in range(100) for n in range(100) if (m**2 + 2**0.5*n**2 < 5000)&(m**2+ 2**0.5*n**2 > 4000)]
```

```python
x = [ m**2 + 2**0.5*n**2 for (m,n) in X]
x = np.array(x)
x = np.sort(x)

T = np.arange(1000)
y = np.zeros(1000)

for t in x:
    y[( T > t - 4000 )] += 1

plt.plot(T + 4000, y, '-')

plt.plot( T+4000, T*np.pi/4/2**0.5)

plt.grid(True)
plt.axis("Equal")
plt.xlim((4000,5000))
```

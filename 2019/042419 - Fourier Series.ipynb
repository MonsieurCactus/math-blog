```python
S = lambda x : ( x + 2**0.5 ) % 1

T = lambda x : ( x + 3**0.5 ) % 1

dt = 0.001
t = np.arange(0, 1, dt)

f = lambda x : 1*(x < 0.5) + 0*(x > 0.5)

z = 0*t

M,N = 5,5

for m in range(M):
    for n in range(N):
        z += f( (t + m*2**0.5 + n*3**0.5) % 1 ) 
        
z = 1.0*z/M/N

plt.plot(t, z, '-')

plt.plot(t[(z > 0.51)], 0.05 + 0*z[(z > 0.51)], '.', markersize=0.5)

plt.plot(t, 0*z + 0.51, '-')

plt.plot(t, 100*(np.cumsum(z)*dt - 0.5*t), '-', linewidth=0.5)

plt.ylim(0,1)
plt.grid(True)
```

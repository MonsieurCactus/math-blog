```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 5,5
```

```python
dt = 0.01
t = np.arange(0,1+dt,0.01)
s = np.arange(0,1,0.05 ) 
```

```python
z = np.exp(2j*np.pi*t)
w = np.exp(2j*np.pi*s)
```

```python
f = z[...,None] + w[None,...]
```

```python
plt.plot(f.real, f.imag, '-')
plt.axis("Equal")
plt.show()
```

```python
dt = 0.01
t = np.arange(0,1+dt,0.01)
s = np.arange(0,1,0.025 ) 

z = np.exp(2j*np.pi*t)
w = np.exp(2j*np.pi*s)

f = z[...,None] + 1.25*w[None,...]
plt.plot(f.real, f.imag, 'g-', linewidth=0.25)
plt.axis("Equal")
plt.show()
```

```python
# Farris Curve "wheels on wheels on wheels"

dt = 0.001
t = np.arange(0,1+dt,dt)
ds = 0.05
s = np.arange(0,1+ds,ds) 

z = np.exp(2j*np.pi*t)
w = np.exp(2j*np.pi*s)

f = (z + (1.0/2)*z**7 + 1*(1.0j/3)*np.conj(z)**17)

plt.plot(f.T.real, f.T.imag, '-', linewidth=1)
plt.axis("Equal")
plt.show()
```

```python
# CONSERVATIVE
# https://www.google.com/search?tbm=isch&q=how+to+draw+with+spirograph+1967

dt = 0.001
t = np.arange(0,1+dt,dt)
ds = 0.05
s = np.arange(0,1+ds,ds) 

z = np.exp(2j*np.pi*t)
w = np.exp(2j*np.pi*s)

f = (z +  0.5*z**5)[None,...] + 0.2*w[...,None]

plt.plot(f.T.real, f.T.imag, 'g-', linewidth=0.25)
plt.axis("Equal")
plt.show()
```

# Weather for One Year

```python
import requests
import datetime
from bs4 import BeautifulSoup
import simplejson as json
import itertools

start = [ "0101", "0201", "0301", "0401", "0501", "0601", "0701", "0801", "0901", "1001", "1101", "1201"]
end = ["0131", "0228", "0331", "0430", "0531", "0630", "0731", "0831", "0930", "1031", "1130", "1231"]

A = ["2018%s"%(month) for month in start[4:]] + ["2019%s"%(month) for month in start[:-4]]
B = ["2018%s"%(month) for month in end[4:]  ] + ["2019%s"%(month) for month in end[:-4]  ]

months = [ (A[n],B[n]) for n in range(12) ]

weather = []

for n in range(12):
    print(n)
    #print(months[n])

    headers = {
        'Accept': 'application/json, text/plain, */*',
        'Referer': 'https://www.wunderground.com/history/daily/us/ny/bronxville/KHPN/date/2019-4-01',
        'Origin': 'https://www.wunderground.com',
        'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36',
    }

    params = (
        ('apiKey', '6532d6454b8aa370768e63d6ba5a832e'),
        ('startDate', months[n][0]),
        ('endDate'  , months[n][1]),
        ('units', 'e'),
    )

    response = requests.get('https://api.weather.com/v1/geocode/41.04/-73.76/observations/historical.json', headers=headers, params=params)

    data = json.loads(response.content)
    
    #print(data)

    weather += [ { "time1": x['expire_time_gmt'] , "time2": x['valid_time_gmt'] , "temp": x['temp']} for x in data['observations']]

weather
```

## Hourly Weather Data for a Month

```python
import requests
import datetime
from bs4 import BeautifulSoup
import simplejson as json

headers = {
    'Accept': 'application/json, text/plain, */*',
    'Referer': 'https://www.wunderground.com/history/daily/us/ny/bronxville/KHPN/date/2019-4-01',
    'Origin': 'https://www.wunderground.com',
    'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36',
}

params = (
    ('apiKey', '6532d6454b8aa370768e63d6ba5a832e'),
    ('startDate', '20180401'),
    ('endDate', '20180501'),
    ('units', 'e'),
)

response = requests.get('https://api.weather.com/v1/geocode/41.04/-73.76/observations/historical.json', headers=headers, params=params)

data = json.loads(response.content)
    
weather += [ { "time1": x['expire_time_gmt'] , "time2": x['valid_time_gmt'] , "temp": x['temp']} for x in data['observations']]

weather
```

```python
Temp = np.array([x['temp'] for x in weather])
Temp
```

```python
Time = np.array([x['time1'] for x in weather])
Time
```

```python
import datetime

#x = datetime.datetime.fromtimestamp(t)
#"%s-%s-%s %s:%s"%(x.year, x.month, x.day, x.hour, x.minute)

# F: DateTime -> (String, String)
def F(t):
    x = datetime.datetime.fromtimestamp(t)
    if x.hour < 2:
        return ""
    else:
        return "%s-%s-%s"%(x.year, x.month, x.day)
```

```python
D = [ F(x['time1']) for x in weather]
D
```

```python
x = np.exp(2j*np.pi*0.25*(np.arange(4) + 0.1*np.random.random(4)))
path = np.cumsum([x[k] for k in ((np.random.random(1000)*4)//1).astype(int)])/10
plt.plot(path.real, path.imag, 'k-')

t = np.arange(0,1.01, 0.01)
z = np.exp(2j*np.pi*t)
plt.plot(z.real, z.imag, '-')

plt.axis("Equal")
plt.grid(True)
```

```python
x = np.exp(2j*np.pi*(  0.125*(np.arange(5)-2) + 0.05*(np.random.random(5) - 0.5) ))
print(x)
args = ((np.random.random(10)*5)//1).astype(int)
print(args)
path = np.cumsum([x[k] for k in args ])/10
plt.plot(path.real, path.imag, 'k-')
plt.plot(path.real, path.imag, 'k.')

t = np.arange(0,1.01, 0.01)
z = np.exp(2j*np.pi*t)

for pt in path:
    plt.fill((pt + 0.075*z).real, (pt + 0.075*z).imag, '-', alpha=0.125)

plt.axis("Equal")
plt.grid(True)
```

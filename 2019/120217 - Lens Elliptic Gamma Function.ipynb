```python
t = np.arange(0,1, 0.001)

R,r,d = 1,1.0/7,3

z = (R-r)*np.exp(2j*np.pi*t) + d*np.exp(2j*np.pi*t*(R-r)/r)

plt.plot(z.real, z.imag, '-', markersize=3)

plt.grid(True)
```

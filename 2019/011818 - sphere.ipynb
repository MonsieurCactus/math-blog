```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dt = 0.001
theta = 2*np.pi*np.arange(0,1+dt, dt)
r = 1

p,q = 3,5

x = r*np.cos(p*theta)*np.cos(q*theta)
y = r*np.cos(p*theta)*np.sin(q*theta)
z = r*np.sin(p*theta)
```

```python
A = [[1,0,0],[0,3.0/5,4.0/5],[0,-4.0/5,3.0/5]]

(x,y,z) = np.dot( A, np.vstack((x,y,z)))
```

```python
w = np.random.random((2500,3))
w = w/(np.sum(w**2, axis=1)**0.5)[...,None]
```

```python
plt.rcParams['figure.figsize'] = 6.25, 6.25
```

```python
plt.plot(x,y, color="#2F25A1")

for (a,b) in [(1,1),(-1,1),(1,-1),(-1,-1)]:
    plt.plot(a*w[:,0], b*w[:,1], '.', color='#E8A220', markersize=2)

plt.axis("Equal")
plt.grid(True)
plt.xticks([-1,-0.5,0,0.5,1], ["" for t in range(5)])
plt.yticks([-1,-0.5,0,0.5,1], ["" for t in range(5)])

plt.show()
```

```python
plt.rcParams['figure.figsize'] = 6.5, 6.5
```

```python
dt = 0.001
theta = 2*np.pi*np.arange(0,1+dt, dt)
r = 1

p,q = 3,7

x = r*np.cos(p*theta)*np.cos(q*theta)
y = r*np.cos(p*theta)*np.sin(q*theta)
z = r*np.sin(p*theta)

A = [[1,0,0],[0,3.0/5,4.0/5],[0,-4.0/5,3.0/5]]
(x,y,z) = np.dot( A, np.vstack((x,y,z)))

w = np.random.random((2500,3))
w = w/(np.sum(w**2, axis=1)**0.5)[...,None]

plt.plot(x,y, color="#2F25A1")

for (a,b) in [(1,1),(-1,1),(1,-1),(-1,-1)]:
    plt.plot(a*w[:,0], b*w[:,1], '.', color='#E8A220', markersize=1.5)

plt.axis("Equal")
plt.grid(True)
plt.xticks([-1,-0.5,0,0.5,1], ["" for t in range(5)])
plt.yticks([-1,-0.5,0,0.5,1], ["" for t in range(5)])

plt.show()
```

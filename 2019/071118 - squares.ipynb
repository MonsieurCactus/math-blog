```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 6,6
```

```python
# three line segments at "random" inside of a square

x = np.random.random((4,4))
x
```

```python
x[0,:2], x[0,2:]
```

```python
for t in range(4):
    plt.plot(x[t,:2],x[t,2:], '-')
    
plt.xlim([0,1])
plt.ylim([0,1])
```

```python
p,q = 37,41

t = np.arange(0,1,1.0/37/41)

plt.plot( (p*t[::5])%1 , (q*t[::5])%1 , '-' )
```

```python
p,q = 37,41

t = np.arange(0,1,1.0/37/41)

for delta in np.arange(0,1,0.1)/41:
    plt.plot( (p*(t+delta))%1 , q*(t+delta)%1 , 'k.' , markersize=0.2)
    
plt.plot( (t[::37]*5)%1 , (t[::37])%1 , 'b.' , markersize = 7 )

plt.grid(True)
#plt.title(    "$\\frac{1}{41} ( 5\mathbb{Z} \oplus \mathbb{Z})/(\mathbb{Z} \oplus \mathbb{Z})$\n",fontsize=20)
plt.xticks(1.0/5*np.arange(5), ["" for x in range(5)])
plt.yticks(1.0/5*np.arange(5), ["" for x in range(5)])
plt.show()
```

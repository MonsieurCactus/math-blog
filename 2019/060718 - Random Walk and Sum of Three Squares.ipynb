```python
from fractions import gcd
```

```python
def gcd(a, b):
    """Calculate the Greatest Common Divisor of a and b.

    Unless b==0, the result will have the same sign as b (so that when
    b is divided by it, the result comes out positive).
    """
    while b:
        a, b = b, a%b
    return a
```

```python
X = [(m,n) for m in range(15) for n in range(15) if m**2 + n**2 < 2*100 and m != 0 and n != 0 and gcd(m,n) == 1]

f = lambda (x,y) : ( (x**2 - y**2)*1.0/(x**2+y**2) , (2*x*y)*1.0/(x**2+y**2))
                  
Y = [f(x) for x in X[1:]]

Z = Y + [(-a,-b) for (a,b) in Y]

for (a,b) in [(20*a,20*b) for (a,b) in Z]:
    print "\draw[color=black!20!white, line width=0.3mm] (0,0)--(%s,%s);"%(a,b)
```

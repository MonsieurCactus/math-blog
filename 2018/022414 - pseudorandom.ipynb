```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = 5.5,5.5
```

```python
for N in 1 + np.arange(5):

    x = [( (b*np.sqrt(2) + c*np.sqrt(3))%1, 0) for b in range(-N,N+1) for c in range(-N,N+1)]
    x = np.array(x)
    #print x
plt.plot(x[:,0], x[:,1], '.', markersize=1)
    
plt.xlim([-0.5**5, 1+0.5**5])
plt.ylim([-0.5**7, 0+0.5**7])
```

```python
t = np.arange(0,1,0.01)

z = np.sum(np.sort(x[:,0])[None,...]*(t[...,None] > np.sort(x[:,0])[None,...]), axis=1)

#print t.shape, z.shape

plt.plot(t, z, '-')
```

```python
dt = 0.01
t = np.arange(0,1,dt)
z = np.sort (x[:,1]  )

w = 0*(t[None,...] < z[...,None]) + 1*(t[None,...] > z[...,None])
w2 = np.sum(w*dt, axis=0)

plt.plot(t,w2 - t)
```

```python
y = []

f = lambda x : x**2

for N in 1 + np.arange(100):

    x = [(b*np.sqrt(2) + c*np.sqrt(3))%1 for b in range(-N,N+1) for c in range(-N,N+1)]
    x = np.sort(np.array(x))
    
    y += [(N, np.sum(f(x))/(4*N**2) )]
    
y = np.array(y)
```

```python
N = 100

n = np.arange(-N,N+1)

x = (2**0.5*n[None,...]+ 3**0.5*n[...,None])%1
x[100,100] = 1

# plug into (non)-parametric statistics textbook
# 
# all of statistics
# all of non-parameteric statistics
# random fragmentation processes
plt.hist( x.reshape(((2*N+1)**2)) , bins = 200)
plt.show()
```

```python
# this is a step function! 

z = [] 

T = np.arange(1,250)

for N in T:

    n = np.arange(-N,N+1)

    x = (7**0.5*n[None,...]+ 3**0.5*n[...,None])%1
    x[N,N] = 1

    a,b = np.where(x == np.min(x))
    z += [x[a[0],b[0]]]
    
plt.plot(T,np.log(z) + 2*np.log(T), '.') 
```

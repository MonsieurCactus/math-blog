## Nodal Lines

```python
x = np.arange(-1,1, 0.01)
y = x
z = x[None,...] + 1j*x[...,None]

w = np.sin(2*np.pi*z.real) + 2*np.sin(2*np.pi*z.imag)
plt.contour(w)

plt.show()
```

```python
n = 5*13*17

#print np.arange(n**0.5)

F = [(int(a),int(b)) for a in np.arange(n**0.5) for b in np.arange(n**0.5) if a**2 + b**2 == n]
#print F

x = np.arange(0,0.5, 0.00025)

z = x[None,...] + 1j*x[...,None]

a = np.random.normal(-1,1,len(F))
#print a

w1 = np.array( [ np.cos(2*np.pi*(f[0]*z.real+f[1]*z.imag)) for f in F ] )
w1 = np.sum(     a[...,None,None]*w1 , axis=0 )

w2 = np.array( [ np.cos(2*np.pi*(f[0]*z.real-f[1]*z.imag)) for f in F ] )
w2 = np.sum(     a[...,None,None]*w1 , axis=0 )

w = w1 + w2

mask = np.where(w > 0)
#print w.shape

plt.plot( z[mask].real , z[mask].imag ,'.', markersize=0.05)
plt.xticks([])
plt.yticks([])
plt.axis("Off")
plt.show()
```

## Farey Fractions $\overline{\mathbb{Q}} = \mathbb{R}$

```python
# generate Farey Fractions up to size N
#      generate list of ordered pairs
#      take GCD of each pair
#      sort them (are the statistics of this random pair?)
# 
# https://stackoverflow.com/questions/11175131/code-for-greatest-common-divisor-in-python

def gcd(a, b):
    """Calculate the Greatest Common Divisor of a and b.

    Unless b==0, the result will have the same sign as b (so that when
    b is divided by it, the result comes out positive).
    """
    while b:
        a, b = b, a%b
    return a
```

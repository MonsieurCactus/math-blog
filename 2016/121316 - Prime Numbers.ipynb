```python


x = 1.0*np.arange(-1,1,0.1)
y = 1.0*x

z = x[...,None] + 1j*y[None,...]

f = lambda w: np.sin(w)

D = lambda f:  lambda x:  f(x+0.01) - f(x)/(0.01)

#plt.plot(z.real, z.imag, 'b', alpha=0.5)
#plt.plot(z.T.real, z.T.imag, 'b', alpha=0.5)

g = lambda w: D(f)(0.5j)*w*0.005 + f(0.5j)

plt.plot(f(z).real  , f(z).imag, 'b', alpha=0.5)
plt.plot(f(z).T.real, f(z).T.imag, 'b', alpha=0.5)

x = 1.0*np.arange(-2,2.5,0.5)
y = 1.0*x

z = x[...,None] + 1j*y[None,...]

plt.plot(g(z).real  , g(z).imag, 'r', alpha=0.45)
plt.plot(g(z).T.real, g(z).T.imag, 'r', alpha=0.45)

plt.plot([f(0.5j).real], [f(0.5j).imag], 'r.', markersize=5)

plt.axis("Equal")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,8
```

```python
dt = 10**-3
t = np.arange(0,1 + dt,dt)

# draw donut
a, b, = 2, 8
r = 4
X = r*np.cos(a*2*np.pi*t) + np.cos(b*2*np.pi*t)
Y = r*np.sin(a*2*np.pi*t) + np.cos(b*2*np.pi*t)
Z = np.sin(b*2*np.pi*t)

# projection operator
A = np.array([[3.0/5, 4.0/5,0],[-4.0/5, 3.0/5,0],[0,0,1]])
B = np.array([[1,0,0],[0,3.0/5, 4.0/5],[0,-4.0/5, 3.0/5]])

C = np.dot(A,B)

# just write out "matrix multiplication"
x = C[0,0]*X + C[0,1]*Y + C[0,2]*Z
y = C[1,0]*X + C[1,1]*Y + C[1,2]*Z
z = C[2,0]*X + C[2,1]*Y + C[2,2]*Z

plt.plot(x, z, 'k-', linewidth=2)


# ------------------------------------------------

# draw donut

N = 10
for n in range(N):

    a, b, = 1, 1
    r = 4
    X = r*np.cos(a*2*np.pi*t) + np.cos(b*2*np.pi*(n*1.0/N))
    Y = r*np.sin(a*2*np.pi*t) + np.cos(b*2*np.pi*(n*1.0/N))
    Z = np.sin(b*2*np.pi*(n*1.0/N))

    # projection operator
    A = np.array([[3.0/5, 4.0/5,0],[-4.0/5, 3.0/5,0],[0,0,1]])
    B = np.array([[1,0,0],[0,3.0/5, 4.0/5],[0,-4.0/5, 3.0/5]])

    # just write out "matrix multiplication"
    x = C[0,0]*X + C[0,1]*Y + C[0,2]*Z
    y = C[1,0]*X + C[1,1]*Y + C[1,2]*Z
    z = C[2,0]*X + C[2,1]*Y + C[2,2]*Z


    plt.plot(x, z, 'k-', linewidth=0.5)
    
    
# ------------------------------------------------

# draw donut

N = 30
for n in range(N):

    a, b, = 1, 1
    r = 4
    X = r*np.cos(a*2*np.pi*(n*1.0/N)) + np.cos(b*2*np.pi*t)
    Y = r*np.sin(a*2*np.pi*(n*1.0/N)) + np.cos(b*2*np.pi*t)
    Z = np.sin(b*2*np.pi*t)

    # projection operator
    A = np.array([[3.0/5, 4.0/5,0],[-4.0/5, 3.0/5,0],[0,0,1]])
    B = np.array([[1,0,0],[0,3.0/5, 4.0/5],[0,-4.0/5, 3.0/5]])

    # just write out "matrix multiplication"
    x = C[0,0]*X + C[0,1]*Y + C[0,2]*Z
    y = C[1,0]*X + C[1,1]*Y + C[1,2]*Z
    z = C[2,0]*X + C[2,1]*Y + C[2,2]*Z


    plt.plot(x, z, 'k-', linewidth=0.5)

# -------------------------------------------

plt.axis("Equal")
plt.grid(True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10,10
```

```python
dt = 10**-3
x = np.arange(0,1,dt)

z = x[None,...] + 1j*x[...,None]

w = 0*z

for a in np.arange(4):
    for b in np.arange(4):
        # "centered" random variable
        #dz = (1.0/2**3)*(np.random.random() - 0.5) + 1j*(1.0/2**3)*(np.random.random() - 0.5)
        dz = 0.1*(1.0/2**3)*np.exp(2j*np.pi*np.random.random())
        mask =  ( ((z-dz).real - a*1.0/3)**2 + ((z-dz).imag - b*1.0/3)**2 < (1.0/6)**2 )
        w[mask] = 1 - w[mask]
        
for a in np.arange(10):
    for b in np.arange(10):
        # "centered" random variable
        #dz = (1.0/2**5)*(np.random.random() - 0.5) + 1j*(1.0/2**5)*(np.random.random() - 0.5)
        dz = 0.1*(1.0/2**5)*np.exp(2j*np.pi*np.random.random())
        mask =  ( ((z-dz).real - a*1.0/9)**2 + ((z-dz).imag - b*1.0/9)**2 < (1.0/25)**2 )
        w[mask] = 1 - w[mask]


mask = (w == 1)
plt.plot(z[mask].real, z[mask].imag, 'k.', markersize=1)

plt.xlim(0,1)
plt.ylim(0,1)
```

```python
dt = 10**-3
x = np.arange(0,1,dt)

z = x[None,...] + 1j*x[...,None]

w = 0*z

# boundary circle
dt = 10**-3
t = np.arange(0,1+dt,dt)
z0 = np.exp(2j*np.pi*t)

for a in np.arange(4):
    for b in np.arange(4):
        # "centered" random variable
        #dz = (1.0/2**3)*(np.random.random() - 0.5) + 1j*(1.0/2**3)*(np.random.random() - 0.5)
        dz = 0.1*(1.0/2**3)*np.exp(2j*np.pi*np.random.random())
        mask =  ( ((z-dz).real - a*1.0/3)**2 + ((z-dz).imag - b*1.0/3)**2 < (1.0/6)**2 )
        w[mask] = 1 - w[mask]

        mask1 = ( ((z-dz).real - (3*a)*1.0/9)**2 + ((z-dz).imag - (3*b)*1.0/9)**2 < (1.0/6)**2 )
        
        w0 = (a+1j*b)*(1.0/3) + (1.0/6)*(z0 + dz)
        plt.plot(w0.real, w0.imag, 'k-', linewidth=1)
        
        for c in np.arange(-1,2):
            for d in np.arange(-1,2):
                # "centered" random variable
                #dz = (1.0/2**5)*(np.random.random() - 0.5) + 1j*(1.0/2**5)*(np.random.random() - 0.5)
                dz = 0.3*(1.0/2**5)*np.exp(2j*np.pi*np.random.random())
                mask2 = ( ((z-dz).real - (3*a+c)*1.0/9)**2 + ((z-dz).imag - (3*b+d)*1.0/9)**2 < (1.0/25)**2 )
                mask = mask1 & mask2
                w[mask] = 1 - w[mask]
                



mask = (w == 1)
plt.plot(z[mask].real, z[mask].imag, 'k.', markersize=1)



        


plt.xlim(0,1)
plt.ylim(0,1)
```

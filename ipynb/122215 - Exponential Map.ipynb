```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10,5
```

```python

dx = 0.05
x = np.arange(0.25,0.75+dx,dx)
dy = 0.01
y = np.arange(0.25,0.5+dy,dy)

w = np.exp(np.pi*np.hstack((     x + 0.25j, 0.75+1j*y, 0.5j + x[::-1], 0.25 + 1j*y[::-1] )))

plt.fill(w.real, w.imag, color="#A4FC9F")

dx = 0.05
x = np.pi*np.arange(0,1+dx,dx)
dy = 0.005
y = np.pi*np.arange(0,1+dy,dy)

z = x[...,None] + 1j*x[None,...]
w = np.exp(z)

plt.plot(w.real, w.imag, color='red', linewidth=1)
plt.plot(w.T.real, w.T.imag, color='blue', linewidth=1)
plt.show()
```

```python
def myCircle(z,r):
    dt = 0.01
    t = np.arange(0,1+dt,dt)
    return z + r*np.exp(2j*np.pi*t)
```

```python
plt.rcParams['figure.figsize'] = 5,5
```

```python
N = 20
dt = 1.0/N
t = np.arange(0,1, dt)
x = np.hstack(( t, 1+1j*t, 1+1j - t, 1j - 1j*t ))



for k in range(-2*N,2*N):
    w = myCircle( x[k], abs(k)*1.0/3*dt)
    plt.fill(w.real, w.imag, alpha=0.1, color='k')

plt.plot([0.5], [0.5], '.', markersize=10, color='#AAAA00')
plt.plot(x.real, x.imag, '.', color="#0033FF", markersize=10)

plt.plot(w.real, w.imag, color='#00FF22')
    
plt.xlim([-0.125,1.125])
plt.ylim([-0.125,1.125])
```

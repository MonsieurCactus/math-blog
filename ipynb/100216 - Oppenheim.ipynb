```python
def myCircle(z,r):
    t = np.arange(0,1,0.01)
    return z + r*np.exp(2j*np.pi*t)
```

```python
N = 100
n = np.arange(N)

R = np.sqrt(n[...,None]**2 + n[None,...]**2)

r = np.sqrt(R**2/np.sqrt(2))//1

P = R**2 - np.sqrt(2)*r**2
Q = R**2 - np.sqrt(2)*(r+1)**2

z = np.sqrt((P*100).astype(int))

for a in np.arange(N):
    for b in np.arange(N):
        w = myCircle(a + 1j*b, z[a,b]/500)
        plt.fill(w.real, w.imag, 'k')
        
plt.xlim([-1,N])
plt.ylim([-1,N])
```

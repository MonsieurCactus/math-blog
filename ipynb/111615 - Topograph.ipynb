```python
def mycircle(z,r):
    dt = 0.01
    t = np.arange(0, 1+dt,dt)
    return z + r*np.exp(2j*np.pi*t)
```

```python
color = ['#10DD10', '#101010']

ds = 0.1

for s in np.hstack((np.arange(0,1, ds), np.arange(1+ds,2, ds))):

    phi = np.pi*s

    w = mycircle(1.0/np.cos(phi/2), np.tan(phi/2))
    plt.plot(w.real, w.imag, color = color[int(s/ds)% 2])
    
w = 1j*np.arange(-2,2, 0.01)    
plt.plot(w.real, w.imag, '#10DD10')

w = mycircle(0,1)
plt.plot(w.real, w.imag, linewidth=3, color="#1010DD")

plt.axis("Equal")
L = 1.25
plt.xlim([-L,L])
plt.ylim([-L,L])
```

```python
def myArc(z,r, (theta1, theta2)):
    dt = 0.01
    t = np.arange(theta1, theta2+dt,dt)
    return z + r*np.exp(1j*t)
```

```python
def myArc((theta1, theta2)):
    theta = 0.5*(theta1 + theta2)*np.pi/2
    phi   = (theta2 - theta1)*np.pi/2
    
    z = np.exp(1j*theta)*1.0/np.cos(phi/2)
    r = np.tan(phi/2)
    
    dt = 0.01
    t = np.arange(np.pi/2 + phi/2 + theta , theta + 3*np.pi/2 - phi/2 +dt, dt)
    
    return z + r*np.exp(1j*t)

def myDot((theta1, theta2)):
    theta = 0.5*(theta1 + theta2)*np.pi/2
    phi   = (theta2 - theta1)*np.pi/2
    
    z = np.exp(1j*theta)*(1.0/np.cos(phi/2) - np.tan(phi/2))**0.75
    r = 0.005
    
    dt = 0.01
    t = np.arange(0,1+dt, dt)
    
    return z + r*np.exp(2j*np.pi*t)
```

```python
def gcd(a,b):
    while b:
        a,b = b, a% b
    return a
```

```python
f = lambda x, y: x[0]*y[1] - y[0]*x[1]

F = []

X = []

for N in range(2, 5):

    F =  sorted([(a,b) for b in range(1,N) for a in range(b+1) if gcd(a,b) ==1],f)
    
    for t in range(len(F)-1):
        X += [( F[t], F[t+1]  )]
        
X
```

```python
X = np.array([[0,1]])

for N in range(3, 5):
    F = np.unique([a*1.0/b for b in range(1,N) for a in range(b+1)])
    X = np.vstack( (X,np.vstack( (F[:-1], F[1:])).T))
X
```

```python
def myArc((x,y)):
    theta1 = x[0]*1.0/x[1]
    theta2 = y[0]*1.0/y[1]
    
    theta = 0.5*(theta1 + theta2)*np.pi/2
    phi   = (theta2 - theta1)*np.pi/2
    
    z = np.exp(1j*theta)*1.0/np.cos(phi/2)
    r = np.tan(phi/2)
    
    dt = 0.01
    t = np.arange(np.pi/2 + phi/2 + theta , theta + 3*np.pi/2 - phi/2 +dt, dt)
    
    return z + r*np.exp(1j*t)

def myDot(( x, y)):
    
    theta1 = x[0]*1.0/x[1]*np.pi/2
    theta2 = y[0]*1.0/y[1]*np.pi/2
    
    theta = (x[0] + y[0])*1.0/(x[1] + y[1])*np.pi/2
    phi   = (theta2 - theta1)
    
    z  = np.exp(1j*theta)*(1.0/np.cos(phi/2) - np.tan(phi/2))**0.5
    #z = np.exp(1j*theta)*(1.0/np.cos(phi/2) - np.tan(phi/2) + 0.5)/1.5
    r = 0.005
    
    dt = 0.01
    t = np.arange(0,1+dt, dt)
    
    return z + r*np.exp(2j*np.pi*t)
```

```python
plt.rcParams['figure.figsize']=7,7
for (a,b) in X:    
    
    #print a,b
    w = myArc((a,b))
    plt.plot(w.real, w.imag, linewidth=3, color='#C0DD10')
    w = myDot((a,b))
    plt.fill(w.real, w.imag, linewidth=3, color='k')
    
for (c,d) in X:  
    
    
    a,b = [c[0] + c[1], c[1]],[d[0] + d[1], d[1]]
    
    w = myArc((a,b))
    plt.plot(w.real, w.imag, linewidth=3, color='#C0DD10')
    w = myDot((a,b))
    plt.fill(w.real, w.imag, linewidth=3, color='k')
    
for (c,d) in X:  
    a,b = [c[0] - c[1], c[1]],[d[0] - d[1], d[1]]
    
    w = myArc((a,b))
    plt.plot(w.real, w.imag, linewidth=3, color='#C0DD10')
    w = myDot((a,b))
    plt.fill(w.real, w.imag, linewidth=3, color='k')
    
for (c,d) in X:      
    a,b = [c[0] - 2*c[1], c[1]],[d[0] - 2*d[1], d[1]]
    
    w = myArc((a,b))
    plt.plot(w.real, w.imag, linewidth=3, color='#C0DD10')
    w = myDot((a,b))
    plt.fill(w.real, w.imag, linewidth=3, color='k')
    
eps = 0.001
for (a,b) in [((0+eps,1), (2-eps,1)),((-2+eps,1), (0-eps,1))]:    
    w = np.arange(-1,1.01, 0.01)
    plt.plot(w.real, w.imag, linewidth=3, color='#C0DD10')
    w = 0.125*(a[0] + b[0])*1j + 0.005*np.exp(2j*np.pi*np.arange(0,1,0.1))
    plt.fill(w.real, w.imag, linewidth=3, color='k')



w = mycircle(0,1)
plt.plot(w.real, w.imag, linewidth=3, color="#1010DD")

plt.axis("Equal")
L = 1.125
plt.xlim([-L,L])
plt.ylim([-L,L])
```

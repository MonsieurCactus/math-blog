```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 5,5
```

```python
X = []

K = np.arange(0,5, 0.01)

for k in K:
    dt = 0.5**k
    t = np.arange(0,1, dt)
    x = np.sin(2*np.pi*t)

    X += [np.sum(x)*dt]
    
X = np.array(X)
    
#plt.plot( 2**K, X, '.', markersize=3)

plt.plot( 2**K, X, '-')

plt.xlabel("# of intervals $= 2^{-\Delta t}$", fontsize=20)
plt.ylabel("$\sum_{k=0}^{1/\Delta t}\; \sin (2 \pi k \Delta t)\\times \Delta t$", fontsize=20)

#plt.ylim([-1,0.])
```

```python
def gcd(a,b):
    while b:
            a, b = b, a%b
    return a
```

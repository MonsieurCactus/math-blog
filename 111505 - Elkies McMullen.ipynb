# Gap Distribution in $\sqrt{n}$ mod 1

Can you get why $\{ \sqrt{n} \}  = 0$ happens more than any other value?  

* What happens when $n = 9$ or $16$?  
* What happens when $n = 10 = 9 + 1$ or $17 = 16 + 1$ ?

```python
N = 5000

n = np.arange(N)

plt.hist(np.sqrt(n)% 1, bins = 100)
```

We reproduce the beautiful result of Elkies and McMullen of the limit distirbution of **gaps** in $\sqrt{n}$ on the circle $S^1$.

* what do we mean "gaps"?
* what do we mean "on the circle"?

Our procedure is to generate **1 million** data points, sort them, compute gaps by subtracting neighbors and plotting the histogram.  

```python
N = 1000000
n = np.arange(N)

m = np.sort(n**0.5 % 1)

gap = m[1:] - m[:-1]

plt.hist(gap, bins = np.arange(0, 5.0/N, 0.1/N))
```

We verify that for random points the histogram of gap distribution is completely different.  What gives??

```python
N = 1000000
n = np.random.random(N)

m = np.sort(n % 1)

gap = m[1:] - m[:-1]

plt.hist(gap, bins = np.arange(0, 5.0/N, 0.1/N))
```

If instead of $\sqrt{n} = n^{0.5}$ we use $n^{0.01} \sqrt{n} = n^{0.501}$ we get no special pattern.

```python
N = 1000000
n = np.arange(N)

m = np.sort(n**0.5 % 1)

#gap = m[1:] - m[:-1]
#plt.hist(gap, bins = np.arange(0, 5.0/N, 0.1/N))

gap = N*(m[1:] - m[:-1])
plt.hist(gap, bins = np.arange(0, 5.0, 0.05))
```

The first surprise about $\sqrt{n} \% 1$ is the numbers start to behave quite regular.  More regular than you think.

This is because:

$$ \sqrt{n+1} - \sqrt{n}  = \frac{1}{\sqrt{n+1} + \sqrt{n}} \approx \frac{1}{2 \sqrt{n}} $$

The simplest example of this trick is $\sqrt{101} = \sqrt{100} + \frac{1}{2\sqrt{100}} \approx 10 + \frac{1}{20} = \color{red}{10}.\color{red}{05}$ these classic ideas were used to estimate square roots without the assistance of a calculator or smart-phone.

```python
plt.rcParams['figure.figsize']=16,5

N = 5000

n = np.arange(N)

plt.plot(np.sqrt(n)% 1, '.', markersize=2)

plt.xticks(  np.arange(np.sqrt(N))**2, ['*' for t in np.arange(np.sqrt(N))**2])
plt.grid(True)
plt.show()
```

# Ergodic Theory

Elkies and McMullin use **ergodic theory** to prove their result, an exact formula on the distribution of the gaps.

$$ m + x < \sqrt{n} < m + x + \frac{t}{N}$$

The intervals are are trying to check are very small, the size is $\frac{1}{N}$.  If we need to be more precise something like:

> $m \in \mathbb{Z}$, $x \in [0,1] \simeq S^1$ and $t \in \mathbb{R}$, $N \gg 1$. 

Let's take the square of both sides of the equation to see if we have any better luck:

$$ \color{#DDD}{\left(m + x \right)^2< n < \left(m + x + \frac{t}{N}\right)^2 \approx 
\left(m + x \right)^2 + \frac{2t}{N}\left(m + x \right)}$$

The gist of Elkies-McMullen is that $x$ and $t$ can be thought of as a pair $(x,t) \in \mathbb{R}^2$ in the plane.

```python
N = 100000


n = np.arange(N)

x,t = np.random.random(), 1000.0

mask = np.where((x < (np.sqrt(n) % 1)) & ((np.sqrt(n) % 1) < x + t/N))


plt.plot(n[mask], np.sqrt(n[mask]) % 1, '.', markersize=5)


#plt.xlim([3*N/4,N])
plt.ylim([x, x + t/N])

plt.xticks(  np.arange(0,np.sqrt(N),10)**2, ['*' for t in np.arange(0,np.sqrt(N),10)**2])
plt.grid(True)
plt.show()
```

$$ \boxed{\sqrt{n+1} - \sqrt{n}  = \frac{1}{\sqrt{n+1} + \sqrt{n}} \approx \frac{1}{\color{#2E5}{2} \sqrt{\color{#2E5}{n}}} }$$

Our substition may be adequate and we are solving a much easier problem!  How do we know this is good enough? 

$$ \frac{1}{2}\left( 1 + \frac{1}{\sqrt{2}} + \frac{1}{\sqrt{3}} + \dots  + \frac{1}{\sqrt{n}}\right) - \sqrt{n} <  \dots $$
